In order to merge our data into one data frame, we needed to have a column to merge on.
The data from NOAA was needed in a special format (City, State, Country). Ex Austin, TX, US
We needed to add the states for each of our cities in the list, so this notebook was made

```python
import pandas as pd
import numpy as np
from city_list import slist, clist
```

```python
masterlist = pd.DataFrame(clist, columns = ["City"])
masterlist["State"] = slist
masterlist.reset_index()
masterlist
```
