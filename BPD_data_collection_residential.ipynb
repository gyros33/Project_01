```python
def get_value(series):
    return (series[0]["value"])
```

```python
for city in clist:
    payload = {
        "filters":{"building_class":["Residential"],
                   "city":[city]},
        "group_by":["building_class"],
        "analyze_by":"source_eui"
    }
    
    try:
        response = requests.post(url=url,data=json.dumps(payload),headers=headers,verify=True)
        res_json = response.json()
        source_eui.append(res_json["table"][0]["mean"])
        print(f"Processing Record : {city}")
        
    except:
        print(f"Missing Data : {city}")
        source_eui.append(np.nan)
        
```

```python
for city in clist:
    payload = {
        "filters":{"building_class":["Residential"],
                   "city":[city]},
        "group_by":["building_class"],
        "analyze_by":"site_eui"
    }
    
    try:
        response = requests.post(url=url,data=json.dumps(payload),headers=headers,verify=True)
        res_json = response.json()
        site_eui.append(res_json["table"][0]["mean"])
        print(f"Processing Record : {city}")
        
    except:
        print(f"Missing Data : {city}")
        site_eui.append(np.nan)
        
```

```python
energy_data["mean source eui"] = source_eui
energy_data["mean site eui"] = site_eui
energy_data
```

```python
energy_data.to_csv(os.path.join("Data","bpd_data_residential.csv"),index=False)
```
