```python
import pandas as pd
import numpy as np
import requests
import json
import os
from config import username, api_key
```

```python
city_path = os.path.join("Data","usa_cities.csv")
df = pd.read_csv(city_path)
df.head()
```

```python
df_test = df.sort_values(by=["province", "pop"],ascending=False)
```

```python
def my_func(pops):
    return list(pops[:5])
df_city_list = df_test.groupby("province").agg({"pop": my_func, "city": my_func})
```

```python
df_city_list = df_city_list.reset_index()
df_city_list
```

```python
clist = []
for cities in df_city_list["city"]:
    clist.extend(cities)
print(clist)
```

```python
#Save config information
url = "https://bpd-api.lbl.gov/api/v2/analyze/table"

# Build query URL
# params = {
#     "units":"imperial"
# }

energy_data = pd.DataFrame(clist,columns=["City"])
bad_cities = []
site_eui = []
number_of_people = []
operating_hours = []
occupant_density = []
floor_area = []
source_eui = []
num_buildings = []


headers = {
    "Content-Type":"application/json",
    "Authorization":f"ApiKey {username}:{api_key}",
}

energy_data
```

```python
payload = {
        "filters":{"city":["Austin"]},
        "group_by":["building_class"],
        "analyze_by":"source_eui"
    }
    
response = requests.post(url=url,data=json.dumps(payload),headers=headers,verify=True)
res_json = response.json()
res_json
```

```python
res_json["table"][2]
```

```python
for city in clist:
    payload = {
        "filters":{"city":[city]},
        "group_by":["building_class"],
        "analyze_by":"source_eui"
    }
    
    try:
        response = requests.post(url=url,data=json.dumps(payload),headers=headers,verify=True)
        res_json = response.json()
        source_eui.append(res_json["table"][2]["mean"])
        print(f"Processing Record : {city}")
        
    except:
        print(f"Missing Data : {city}")
        source_eui.append(np.nan)
        
```

```python
energy_data["mean source eui"] = source_eui
energy_data
```

```python
energy_data.to_csv(os.path.join("Data","bpd_data.csv"))
```
